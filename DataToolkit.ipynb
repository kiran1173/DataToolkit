# 1. What is NumPy, and why is it widely used in Python?
   - NumPy (Numerical Python) is a powerful Python library used for numerical and scientific computing. It provides support for multi-dimensional arrays, mathematical operations, linear algebra, Fourier transforms, random number generation, and more. Unlike Python lists, NumPy arrays are stored in contiguous memory blocks, making computations much faster. This efficiency makes NumPy the backbone of many other data science libraries such as Pandas, SciPy, Scikit-learn, and TensorFlow.

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
print("Mean:", np.mean(arr))
```

# 2. How does broadcasting work in NumPy?
   - Broadcasting is a mechanism that allows NumPy to perform operations on arrays of different shapes without explicitly copying data. Instead of looping through elements, NumPy expands smaller arrays to match larger ones in shape, making operations faster and memory-efficient.

```python
a = np.array([1, 2, 3])
b = 2
print(a + b)
```

# 4. Explain the use of the groupby() method in Pandas.
   - The groupby() method is used to split data into groups based on column values and then apply aggregation functions (like sum, mean, count). It is commonly used in data analysis to understand patterns and trends.

# 6. What are the differences between NumPy arrays and Python list?
   - Python lists are general-purpose containers that can hold different data types, while NumPy arrays are homogeneous and optimized for numerical operations. NumPy arrays consume less memory, provide vectorized operations, and are much faster.

```python
import time
lst = list(range(1000000))
arr = np.array(lst)

start = time.time()
[x*2 for x in lst]
print("List time:", time.time() - start)

start = time.time()
arr*2
print("Array time:", time.time() - start)
```

# 7. What is a heatmap, and when should it be used?
   - A heatmap is a data visualization that uses color intensity to represent values in a matrix or table. Itâ€™s useful for showing correlations, frequency counts, or intensity levels in data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create sample data
data = {
    'Math': [90, 80, 85, 70, 95],
    'Science': [85, 60, 78, 75, 92],
    'English': [88, 72, 80, 65, 85],
    'History': [70, 65, 75, 60, 80]
}

df = pd.DataFrame(data)

# Compute correlation matrix
corr = df.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

# 9. How does Matplotlib differ from Plotly?
   - Matplotlib and Plotly are both powerful visualization libraries, but they serve different purposes. Matplotlib is a low-level, static visualization library, great for creating research papers, academic plots, and custom designs. It provides full control over every aspect of a plot but often requires more code to make charts visually appealing. On the other hand, Plotly is a high-level, interactive visualization library that allows zooming, panning, tooltips, and dynamic dashboards. Plotly is widely used in web applications and business dashboards because it creates interactive visuals easily

```python
import matplotlib.pyplot as plt
import plotly.express as px

# Matplotlib (static)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Matplotlib Static Plot")
plt.show()

# Plotly (interactive)
fig = px.line(x=[1, 2, 3], y=[4, 5, 6], title="Plotly Interactive Plot")
fig.show()
```

# 10. What is the significance of hierarchical indexing in PandaS?
      

- Hierarchical indexing (MultiIndex) in Pandas allows you to have multiple levels of row or column labels. This feature is significant because it lets you work with higher-dimensional data in a 2D table, making complex data analysis easier. It is especially useful when handling time-series data with multiple categories or when representing grouped data.

```python
import pandas as pd
arrays = [
    ['A', 'A', 'B', 'B'],
    ['Math', 'Science', 'Math', 'Science']
]
index = pd.MultiIndex.from_arrays(arrays, names=('Student', 'Subject'))
df = pd.DataFrame([90, 85, 78, 88], index=index, columns=['Score'])
print(df)
```

# 19. How does Pandas simplify time series analysis?
    

- Pandas has built-in time-series functionality with date parsing, resampling, frequency conversion, and rolling windows. This makes it easy to handle time-indexed data.

```python
import pandas as pd
date_rng = pd.date_range(start='2023-01-01', periods=5, freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['value'] = [10, 20, 30, 40, 50]
df.set_index('date', inplace=True)
print(df.resample('2D').mean())
```

# 20. What is the role of a pivot table in Pandas?
    

- A pivot table summarizes data by aggregating values based on categories. It is similar to Excel pivot tables and is powerful for reporting and data analysis.

```python
df = pd.DataFrame({'Dept':['HR','IT','HR','IT'],
                   'Salary':[3000,4000,3500,4500]})
print(df.pivot_table(values='Salary', index='Dept', aggfunc='mean'))
```

# 21. Why is NumPyâ€™s array slicing faster than Pythonâ€™s list slicing?
    

- NumPy arrays are stored in contiguous memory and slicing creates a view, not a copy, meaning no data duplication occurs. In contrast, Python lists store references to objects and slicing creates a new list.

```python
import numpy as np
arr = np.arange(10)
print(arr[2:7])
```
